```python
import pandas as pd
import os
from itertools import product
```

```python
folder_path = '../data'

# Get datasets
datasets = [
    name for name in os.listdir(folder_path)
    if os.path.isdir(os.path.join(folder_path, name))
]

param_combinations = []

for dataset in datasets:
    test_folds = sorted(
        pd.read_csv(os.path.join(folder_path, dataset, 'folds.csv'))['fold'].unique()
    )

    combinations = product(
        [dataset],
        test_folds,
        [2, 4, 8],          # n_conv_layers
        [2, 4, 8],          # n_filters
        [2, 4, 8],          # kernel_size
        [1, 2],             # n_dense_layers
        [10, 20, 50]        # n_neurons
    )

    param_combinations.extend(combinations)

# Create DataFrame
params_df = pd.DataFrame(
    param_combinations,
    columns=[
        'dataset',
        'test_fold',
        'n_conv_layers',
        'n_filters',
        'kernel_size',
        'n_dense_layers',
        'n_neurons'
    ]
)

# Save to CSV
params_df.to_csv("params_model.csv", index=False)
```
